## Reordering Columns

```python
df.columns
```

```python
order = ['id', 'name', 'long_name', 'age', 'nationality', 'height', 'weight',
         'foot', 'wage', 'value', 'joined']
```

```python
remaining_columns = [col for col in df.columns if col not in order]
```

```python
df = df[order + remaining_columns]
```

```python
df.columns
```

```python
df['team & contract']
```

```python
df["team"] = df["team & contract"].str.extract(r'(\n\n\n\n.*\n)', expand=False).str.strip()
df["contract"] = df["team & contract"].str.findall(r'[\d~ ]+').str[-1]
```

```python
df["contract_start"] = df["contract"].str.slice(0, 4)
df["contract_end"] = df["contract"].str.slice(-4)
```

```python
df["contract_start"] = pd.to_numeric(df["contract_start"], errors="coerce").fillna(0).astype(int)
df["contract_end"] = pd.to_numeric(df["contract_end"], errors="coerce").fillna(0).astype(int)
```

```python
df.drop("team & contract", axis=1, inplace=True)
df.drop("contract", axis=1, inplace=True)
```

## Converting data types

```python
def convert_height_to_cm(height):
    # Remove the double quotes and split by the apostrophe
    feet, inches = height.replace('"', '').split("'")
    # Convert to centimeters (1 foot = 30.48 cm, 1 inch = 2.54 cm)
    total_inches = int(feet) * 12 + int(inches)
    return int(total_inches * 2.54)
```

```python
df['height'] = df['height'].apply(convert_height_to_cm)
```

```python
df["height"]
```

```python
df["weight"] = df["weight"].str.replace("lbs", "", regex=False).astype(int)
```

```python
df = df.rename(columns={"weight": "weight_lbs"})
```

```python
df["weight_lbs"]
```

```python
df["joined"]
```

```python
df['joined'] = pd.to_datetime(df['joined'])
```

```python
df["joined"]
```

```python
df[["wage","value","release_clause"]]
```

```python
def convert_money(value):
    value = value.replace('€', '').replace('K', 'e3').replace('M', 'e6')  # Replace suffixes with scientific notation
    return int(float(value))
```

```python
money_columns = ['wage', 'value', 'release_clause']
for col in money_columns:
    df[col] = df[col].apply(convert_money)
```

```python
df.head()
```

```python
df.rename(columns={
    "wage": "wage_in_euro",
    "value": "value_in_euro",
    "release_clause": "release_clause_in_euro"
}, inplace=True)
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Top 10 Players

```python
top_10_players = df.sort_values(by="value_in_euro", ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_players['name'], top_10_players['value_in_euro'], color='skyblue')
plt.xlabel('Value in Euro (€)', fontsize=12)
plt.ylabel('Player Name', fontsize=12)
plt.title('Top 10 Players by Value in Euro', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to show the highest value at the top
plt.tight_layout()

plt.show()
```

```python
top_10_richest = df.sort_values(by="wage_in_euro", ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_players['name'], top_10_richest['wage_in_euro'], color='skyblue')
plt.xlabel('Wage in Euro (€)', fontsize=12)
plt.ylabel('Player Name', fontsize=12)
plt.title('Top 10 Richest Players', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to show the highest value at the top
plt.tight_layout()

plt.show()
```

## Top 10 Nationalities

```python
country_counts = df['nationality'].value_counts()
```

```python
top_10_countries = country_counts.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_countries.index, top_10_countries.values, color='lightcoral')
plt.xlabel('Number of Players', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Top 10 Countries with Most Players', fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Age Distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.title('Age Distribution of Players', fontsize=14)
plt.tight_layout()
plt.show()
```

## Wage vs. Value

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='wage_in_euro', y='value_in_euro', data=df, color='skyblue', s=100, alpha=0.7)
plt.title('Wage vs Value of Players', fontsize=14)
plt.xlabel('Wage in Euro (€)', fontsize=12)
plt.ylabel('Value in Euro (€)', fontsize=12)
plt.tight_layout()
plt.show()
```

## Top 10 Teams

```python
team_value = df.groupby('team')['value_in_euro'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
team_value.plot(kind='bar', color='skyblue')
plt.title('Top 10 Teams with the Most Player Value', fontsize=14)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Total Player Value in Euro (€)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
